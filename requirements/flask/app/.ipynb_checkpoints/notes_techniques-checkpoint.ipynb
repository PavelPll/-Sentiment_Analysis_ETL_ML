```python
from datetime import datetime
d = {}
for el in customer_orders:
    print(el['order_id'], el['order_date'])
    d[el['order_id']] = datetime.strptime(el['order_date'].split("T")[0], '%Y-%m-%d').timestamp()
print(d)
redis_client.zadd("sorted:orders", d)
redis_client.zrange("sorted:orders", 0, -1, desc=True, withscores=True)
print(redis_client.zscan("sorted:orders"))
redis_client.zrem("sorted:orders", 0, -1)
#{'order_date': '2015-02-15T01:51:00',
# 'order_id': 3089,
```

```python
aggregation = [
       {
           "$group": {
               "_id": "$customer.id",
               #"_id": "$product.category_id",
               #"avg_sales_price_per_customer": { "$avg": "$sales" },
               #"orders" : { "$addToSet": "$id" }
               # "orders" : { "$addToSet": "$product.category_id" }
               "orders" : { "$addToSet": "$id" }
           }
       }
]
list(orders.aggregate(aggregation))
ids_customer = {}
for el in list(orders.aggregate(aggregation)):
    ids_customer[el['_id']] = el['orders']
ids_customer
```

```python
# 9 {<order_id_1>: <quantity_1>, ..., <order_id_n>: <quantity_n>}
# http://localhost:5000/customer/4602/orders/3

quantity = 5

total_quantities = redis_client.json().get("customer:orders", '$..total_quantity')
order_ids = redis_client.json().get("customer:orders", '$..order_id')

try:
    redis_client.hdel("order:quantity", *redis_client.hkeys("order:quantity"))
except:
    pass
for (order_id, total_quantity) in zip(order_ids, total_quantities):
    if int(total_quantity)>quantity:
        print(int(total_quantity))
        redis_client.hset("order:quantity", mapping={order_id: total_quantity})
print(redis_client.hgetall("order:quantity"))
```
